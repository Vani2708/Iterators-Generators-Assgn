1.  What is the difference between a function and a method in Python?
  * Function: A function is a block of reusable code that performs a specific task. It is defined using the def keyword and is independent of any object.Functions are called using their name followed by parentheses, e.g., function_name(arguments). Functions are general-purpose callable objects.

  * Methods: A method is a function that is associated with an object. It is defined within a class and can operate on the object (or class) it belongs to. Methods are called on an object using the dot (.) notation, e.g., object.method(arguments). Methods are functions that are bound to an object (instance method) or a class (class/static method).

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
```

```python
class Greeter:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

greeter = Greeter("Alice")
print(greeter.greet())
```

  * Arguments: Arguments are the actual values that are passed to the function when it is called.When the greet function is called with the argument "Alice", the value "Alice" is assigned to the parameter name.

```python
def greet(name):  # `name` is a parameter
    print(f"Hello, {name}!")
```

  2. Default Arguments
  3. Keyword Arguments
  4. Positional Arguments
  5. Lambda functions
  6. Using 'args' and 'kwargs' Together

```python
1. def greet(name):
    print(f"Hello, {name}!")

greet("Alice")  # Function call

2. def greet(name="Guest"):
    print(f"Hello, {name}!")

greet()
greet("Alice")

3. def introduce(name, age):
    print(f"My name is {name} and I am {age} years old.")

introduce(age=25, name="Alice")

4. def add(a, b):
    return a + b

result = add(5, 3)  # Positional arguments
print(result)

5. square = lambda x: x ** 2
print(square(4))


6. def process_data(*args, **kwargs):
    print("Positional:", args)
    print("Keyword:", kwargs)

process_data(10, 20, 30, name="John", age=25)
```

Q4.  What is the purpose of the `return` statement in a Python function?

Q5.  What are iterators in Python and how do they differ from iterables?
    1. Iterables: An iterable is any object capable of returning its members one at a time. Examples include lists, tuples, dictionaries, sets, and strings. These objects implement the __iter__() method, which returns an iterator.
     Examples: Lists, tuples, dictionaries, strings, sets, etc.
    2. Iterators: An iterator is an object that represents a stream of data. It is returned by calling the iter() method on an iterable. Iterators implement both the __iter__() and __next__() methods.

```python
# Iterable
my_list = [10, 20, 30]  # This is an iterable

# Get an iterator from the iterable
my_iterator = iter(my_list)

# Using the iterator to fetch items
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
# print(next(my_iterator))
```

Q7. Explain the concept of generators in Python and how they are defined.
  * Generators are a type of iterable in Python that allow you to yield values one at a time, rather than generating all values at once and storing them in memory. They are particularly useful when working with large datasets or streams of data because they help conserve memory and improve performance. Generators are defined using functions with the yield statement instead of return. When a generator function is called, it returns a generator object, which can be iterated over using a for loop, next(), or other iteration mechanisms.

```python
def simple_generator():
    yield 1
    yield 2
    yield 3

gen = simple_generator()
print(next(gen))
print(next(gen))
print(next(gen))
```

Q8. What are the advantages of using generators over regular functions?
  * The advantages of using generators over regular functions are:
     1. Memory efficiency: Return all data at once, which can require significant memory if the dataset is large.
     2. Lazy evaluation: Values are computed only when requested. This delays computation until it's actually needed, optimizing performance.
     3. Simplicity in writing iterators: utomatically handle the state and logic for iteration. No need to manually implement __iter__() and __next__() methods.
     4. Infinite sequences: Easily create infinite sequences without running out of memory.
     5. Improved performance: Generators donâ€™t require building and returning large data structures like lists or dictionaries, making them faster in scenarios where all data is not needed immediately.
     6. Pipelining: Generators allow data to be processed in a pipeline, passing data from one generator to the next without intermediate storage.
     7. Readability & Maintainability: Generators often result in cleaner, more concise code, especially for tasks involving complex iteration or streaming data.

Q9.  Explain the purpose and usage of the `map()` function in Python.
  * The map() function in Python is used to apply a given function to each item in an iterable (such as a list, tuple, or string) and return a map object, which is an iterator. It is particularly useful for transforming or processing all elements in a collection without using explicit loops.



```python
Using map() with Built-in Functions
You can use map() directly with built-in functions like str, int, len, etc.

str_numbers = ["1", "2", "3"]
int_numbers = map(int, str_numbers)
print(list(int_numbers))
```

```python
Using map() with Multiple Iterables

nums1 = [1, 2, 3]
nums2 = [4, 5, 6]
sums = map(lambda x, y: x + y, nums1, nums2)
print(list(sums))
```

Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  *The map(), reduce(), and filter() functions in Python are part of the functional programming toolkit, allowing concise and efficient transformations and processing of iterables.
  1. map(): Transforms each element in an iterable by applying a given function and returns an iterator with the transformed elements.Use when you want to transform or modify all elements of an iterable.

  2. filter(): Filters elements from an iterable based on whether they satisfy a condition defined by a function.Use when you want to filter out unwanted elements from an iterable based on a condition.

  3. reduce(): Applies a function cumulatively to the elements of an iterable, reducing them to a single value.Use when you need to aggregate or combine all elements of an iterable into a single result.
